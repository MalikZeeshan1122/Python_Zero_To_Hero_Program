```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
X = np.linspace(-5, 5, 100)  

Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
```

```python
Z = X**2 + Y**2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')  
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z = X^2 + Y^2')
plt.show()
```

```python
X = np.linspace(-2 * np.pi, 2 * np.pi, 100)
Y = np.linspace(-2 * np.pi, 2 * np.pi, 100)
X, Y = np.meshgrid(X, Y)
```

```python
Z = X**2 - Y**2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')  
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z = X^2 - Y^2')
plt.show()
```

```python
X = np.linspace(-np.sqrt(23), np.sqrt(23), 100)
Y = np.linspace(-np.sqrt(23), np.sqrt(23), 100)
X, Y = np.meshgrid(X, Y)
```

```python
Z = X**3 + Y**3
```
